```python
from ultralytics import YOLO
import cv2
import cvzone
```

```python
vedio = cv2.VideoCapture(0)
face_detection = YOLO('yolov8n-face.pt')
while True:
    ret,frame = vedio.read()
    
    face_result = face_detection.predict(frame,conf = 0.4)
    for results in face_result:
        parameter = results.boxes
        for box in parameter:
            x1,y1,x2,y2 = box.xyxy[0]
            x1,y1,x2,y2 = int(x1),int(y1),int(x2),int(y2)
            h,w = (y2-y1),(x2-x1)
            rec = cvzone.cornerRect(frame,[x1,y1,w,h],l=9,rt=3,colorR=(0,255,0))
            conf = box.conf[0]
            cvzone.putTextRect(frame,f'Face:{conf:.2f}%',(x1,y1-10),scale=1,thickness=1,colorR=(0,0,0))
    cv2.imshow('frame',frame)
    if cv2.waitKey(40) & 0xff == ord(' '):
        vedio.release
        cv2.destroyAllWindows()
        break
```
