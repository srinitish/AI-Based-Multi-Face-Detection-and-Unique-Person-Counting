```python
import cv2
import cvzone
from ultralytics import YOLO

vedio = cv2.VideoCapture(0)
face_detection = YOLO('yolov8n-face.pt')
unique = set()
while True:
    ret,frame = vedio.read()
    result = face_detection.track(frame,persist=True,conf = 0.4)
    for face in result:
        parameter = face.boxes
        count = len(parameter)
        for box in parameter: 
            x1,y1,x2,y2 = box.xyxy[0]
            conf  = box.conf[0]
            h,w = (y2-y1),(x2-x1)
            x1,y1,w,h = int(x1),int(y1),int(w),int(h)
            cvzone.putTextRect(frame,text=f'Person {conf:.2f}%',pos=(x1,y1-10),scale=1,thickness=1,colorR=(0,0,0))
            cvzone.putTextRect(frame,text=f'No of person {count}',pos=(10,20),scale=1,thickness=1)
            cvzone.cornerRect(frame,[x1,y1,w,h],l=9,rt=3,colorR=(0,255,0))
    cv2.imshow('frame',frame)
    if cv2.waitKey(1)&0xff==ord(' '):
        break
vedio.release()
cv2.destroyAllWindows()

```
